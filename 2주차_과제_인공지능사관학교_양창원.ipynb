### 03번 문제 (slice string)

__다음 변수 date를 슬라이스해서 year(년), month(월), day(일)를 각각 출력하세요.__ (출력 : 2020 05 11)

(tip: None을 지우고 date를 슬라이싱해서 각각의 데이터를 저장하세요!)

```
# 주어진 변수(date)만 직접 수정불가, 다른 year, month, day 변수는 직접 수정가능
# 슬라이스를 사용 안 할 경우 오답처리
date = '20200511'

year = date[0:4] # date의 데이터에서 0번 이상 4번 미만의 데이터 추출
month = date[4:6] # date의 데이터에서 4번 이상 6번 미만의 데이터 추출
day = date[6:8] # date의 데이터에서 6번 이상 8번 미만 데이터 추출

print(year, month, day) # 출력
```

### 08번 문제(산술 연산 및 list index)

__직사각형의 대각선의 길이를 구하는 공식은 밑변의 제곱 + 높이의 제곱 = 대변의 제곱이다.__

__다음 변수 list_08에는 하나의 직사각형의 밑변과 높이가 저장되어 있습니다.__

__각각 밑변이 3이고 높이가 4일 때, 직사각형의 대변의 제곱의 길이를 출력하세요.__

```
# 주어진 변수(list_08) 직접 수정불가
list_08 = [3, 4]

x = list_08[0] # 직사각형 밑변의 길이는 x로 정의
y = list_08[1] # 직사각형 높이의 길이는 y로 정의

z1 = (x**2 + y**2) # 직사각형의 대각선의 제곱의 길이는 z1로 정의, z1 = (x^2 + y^2)
z2 = (x**2 + y**2)**(0.5) # 직사각형의 대각선의 길이는 z2로 정의, z2 = (x^2 + y^2)^(0.5)

print(f"직사각형의 대변의 제곱의 길이 : {z1}") # 대각선의 제곱의 길이 출력
print(f"직사각형의 대변의 길이 : {z2}") # 대각선의 길이 출력
```

### 09번 문제(소수 판별 프로그램)
__입력받은 숫자가 소수인지 아닌지 출력하는 프로그램을 작성 해 주세요.__ (출력 : __소수__ or __소수가 아닙니다__)

```
# 주어진 변수(num_10) 직접 수정불가
# 함수(prime_number) 직접 수정가능
num_10 = int(input('숫자를 입력해주세요: '))

def prime_number(n): #n값을 입력 받아 소수인지 아닌지 구분하는 함수
  if n != 1: # 1은 소수가 아니다. 2, 3, 4, ..., (n-1)까지의 인수에 대해서
    for f in range(2, n): # n이 위의 인수(2, 3, 4, ..., (n-1))의 하나로 나누어지면(나머지 = 0)
      if n % f == 0: 
        return False # 나머지가 0이라면 소수가 아님
  else:              # n이 1이라면
    return False     # 소수가 아님

  return True        # n이 1이 아니면서 2부터 (n-1)까지의 수로 나누어지지 않으면 소수

if prime_number(num_10) == True:
  print("소수")
else:
  print("소수가 아닙니다")
```
